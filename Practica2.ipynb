```python
#Evelyn Sofia y Diego Rene
#EJERCICIO 1

import cv2
import numpy as np
import matplotlib.pyplot as plt


# Cargar la imagen en escala de grises
imagen = cv2.imread("/content/mano.jpg", cv2.IMREAD_GRAYSCALE)

# Dimensiones de la imagen
alto, ancho = imagen.shape


# Definir la matriz de traslacin (50 y 30)
tx, ty = 50, 30
M = np.float32([[1, 0, tx], [0, 1, ty]])

# Aplicar la transformacin afn
imagen_trasladada = cv2.warpAffine(imagen, M, (ancho, alto))


# Mostrar imgenes
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(imagen, cmap="gray")
plt.title("Normal")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(imagen_trasladada, cmap="gray")
plt.title("Trasladada")
plt.axis("off")

plt.show()
```

```python

# Definir matriz de traslacin con valores decimales
tx, ty = 20.5, 15.5  # Desplazamiento en X e Y con decimales
M = np.float32([[1, 0, tx], [0, 1, ty]])

# Aplicar la traslacin con interpolacin
imagen_trasladada = cv2.warpAffine(imagen, M, (ancho, alto), flags=cv2.INTER_LINEAR)


# Mostrar imgenes
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(imagen, cmap="gray")
plt.title("Normal")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(imagen_trasladada, cmap="gray")
plt.title("Trasladada (Decimales)")
plt.axis("off")

plt.show()
```

```python
#EJERCICIO 2


# Definir rotacin
angulo = 45
centro = (ancho // 2, alto // 2)  # Centro de la imagen
escala = 1.0  # Sin cambios en el tamao

M_rotacion = cv2.getRotationMatrix2D(centro, angulo, escala)

# Aplicar rotacin sobre la imagen trasladada
imagen_final = cv2.warpAffine(imagen_trasladada, M_rotacion, (ancho, alto))


# Mostrar imgenes
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.imshow(imagen, cmap="gray")
plt.title("Normal")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(imagen_trasladada, cmap="gray")
plt.title("Trasladada")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(imagen_final, cmap="gray")
plt.title(" Trasladada y Rotada")
plt.axis("off")

plt.show()
```

```python
#EJERCICIO 3

# Definir rotacin + escalado (45 alrededor del centro con 0.8 de escala)
angulo = 45
escala = 0.5
centro = (ancho // 2, alto // 2)  # Centro de la imagen

M_rotacion_escala = cv2.getRotationMatrix2D(centro, angulo, escala)

# Aplicar rotacin y escalado
imagen_50 = cv2.warpAffine(imagen_trasladada, M_rotacion_escala, (ancho, alto))

# Definir rotacin + escalado (45 alrededor del centro con 0.8 de escala)
angulo = 45
escala = 1.5
centro = (ancho // 2, alto // 2)  # Centro de la imagen

M_rotacion_escala = cv2.getRotationMatrix2D(centro, angulo, escala)

# Aplicar rotacin y escalado
imagen_150 = cv2.warpAffine(imagen_trasladada, M_rotacion_escala, (ancho, alto))

# Mostrar imgenes
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.imshow(imagen, cmap="gray")
plt.title("Imagen Original")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(imagen_50, cmap="gray")
plt.title("Imagen 50")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(imagen_150, cmap="gray")
plt.title("Imagen 150")
plt.axis("off")

plt.show()
```

```python
# Cargar la imagen en escala de grises
imagen = cv2.imread("/content/cerebro.jpg", cv2.IMREAD_GRAYSCALE)

# Aplicar ecualizacin de histograma
imagen_ecualizada = cv2.equalizeHist(imagen)

# Calcular histogramas antes y despus
hist_original = cv2.calcHist([imagen], [0], None, [256], [0,256])
hist_ecualizada = cv2.calcHist([imagen_ecualizada], [0], None, [256], [0,256])

# Mostrar imgenes y histogramas
plt.figure(figsize=(10,5))

# Imagen Original
plt.subplot(2,2,1)
plt.imshow(imagen, cmap="gray")
plt.title("Original")
plt.axis("off")

# Histograma Original
plt.subplot(2,2,2)
plt.plot(hist_original, color="black")
plt.title("Histograma Original")
plt.xlabel("Intensidad")
plt.ylabel("Frecuencia")

# Imagen Ecualizada
plt.subplot(2,2,3)
plt.imshow(imagen_ecualizada, cmap="gray")
plt.title("Ecualizada")
plt.axis("off")

# Histograma Ecualizado
plt.subplot(2,2,4)
plt.plot(hist_ecualizada, color="black")
plt.title("Histograma Ecualizado")
plt.xlabel("Intensidad")
plt.ylabel("Frecuencia")

plt.tight_layout()
plt.show()
```
